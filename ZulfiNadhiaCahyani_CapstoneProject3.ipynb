In this environment, **apartment owners**, whether individuals or companies, face significant challenges when setting prices. #### **II.2.4 Handling Inconsistent Data**

**Inconsistent data** refers to data that does not align with expected ranges, formats, or rules, which can **cause errors or discrepancies in analysis**. These inconsistencies often arise due to human error, system glitches, or data entry mistakes. If left unaddressed, such inconsistencies can lead to incorrect results, unreliable insights, and invalid conclusions. In the context of building a regression model, addressing these inconsistencies is crucial to ensure the model produces accurate and reliable predictions. Several inconsistencies have been detected in the `TimeToSubway` column:

1. **Inconsistent formatting:**  
   - One value (`0-5min`) differs from others that use the `~` symbol to represent time ranges. Standardizing these values is essential for the model to interpret the data correctly and avoid errors during preprocessing or model fitting.

2. **Misleading category naming:**  
   - The category `no_bus_stop_nearby` should be renamed to `no_subway_station_nearby` for clarity, as it refers to proximity to subway stations, not bus stops. This correction ensures that the feature is properly understood by both the model and users, avoiding potential misinterpretation.


```python
# CHECK INCONSISTENT DATA

# 1. Check for Inconsistent Formatting in 'TimeToSubway'
inconsistent_time_format = df[df['TimeToSubway'] == '0-5min']
print("Number of rows with inconsistent time formatting (0-5min):", inconsistent_time_format.shape[0])
print(inconsistent_time_format[['TimeToSubway']].head(5))

# 2. Check for Misleading Category Naming in 'TimeToSubway'
misleading_category = df[df['TimeToSubway'] == 'no_bus_stop_nearby']
print("\nNumber of rows with misleading category 'no_bus_stop_nearby':", misleading_category.shape[0])
print(misleading_category[['TimeToSubway']].head(5))
```

```python
# HANDLING INCONSISTENT DATA

# 1. Fix Inconsistent Formatting in 'TimeToSubway'
# Replace '0-5min' with '0~5min' to standardize formatting
df['TimeToSubway'] = df['TimeToSubway'].replace('0-5min', '0~5min')


# 2. Rename Misleading Category 'no_bus_stop_nearby'
# Replace 'no_bus_stop_nearby' with 'no_subway_station_nearby'
df['TimeToSubway'] = df['TimeToSubway'].replace('no_bus_stop_nearby', 'no_subway_station_nearby')
```

#### **II.2.5 Handling Outliers**

**Outliers** are data points that deviate significantly from the overall pattern of the dataset. These are **extreme values** that fall well outside the typical range of the data. Outliers can arise due to various causes, including **data entry errors**, **measurement inaccuracies**, or **natural variation** within the population.

In the context of **regression modeling**, outliers can have a **disproportionate influence** on the model’s parameters, potentially leading to **biased coefficients**, **distorted predictions**, and **reduced model performance**. Identifying and appropriately handling outliers is therefore essential to ensure that the model learns from **representative and reliable data**.


```python
# Check for Numerical Columns
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns
print(numerical_columns)
```

```python
# CHECK OUTLIERS USING IQR

# Create a function to check outliers
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)  # 25th percentile (lower quartile)
    Q3 = df[column].quantile(0.75)  # 75th percentile (upper quartile)
    IQR = Q3 - Q1  # Interquartile Range
    lower_bound = Q1 - 1.5 * IQR  # Lower bound for outliers
    upper_bound = Q3 + 1.5 * IQR  # Upper bound for outliers
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    
    return outliers, lower_bound, upper_bound

# Initialize an empty list to store columns that have outliers
columns_with_outliers = []

# Detect outliers in each numerical column
for col in numerical_columns:
    outliers, lower, upper = detect_outliers_iqr(df, col)
    
    print(f"Column: {col}")
    print(f"Lower Bound: {lower}, Upper Bound: {upper}")
    print(f"Number of Outliers: {outliers.shape[0]}")
    print("-" * 50)
    
    if outliers.shape[0] > 0:
        columns_with_outliers.append(col)

# Print all columns that contain outliers
if columns_with_outliers:
    print("\nThe columns with outliers are:", ", ".join(columns_with_outliers))
else:
    print("\nNo columns contain outliers.")
```

### **II.4  Data Preprocessing**

**Data preprocessing** prepares the dataset for modeling by ensuring it is in a format suitable for machine learning algorithms. It includes **'Data Transformation'**, which involves **scaling** numeric features for uniform treatment and **encoding** categorical data into a **numerical format**, and **'Feature Engineering'**, which creates new features, such as **binning** continuous variables into discrete categories, to enhance model performance.

Combining these techniques optimizes the data, simplifying it while preserving important patterns, leading to improved **model accuracy** and **effectiveness** during training.

<br>

#### **II.4.1 Data Transformation**

**Data transformation**, as part of the broader **data preprocessing** phase, is a vital step that ensures features are in a format suitable for machine learning algorithms. It includes techniques like **scaling** and **encoding**, which enhance model performance by standardizing inputs and converting non-numeric data into usable numerical values. Proper transformation not only improves the **consistency** and **comparability** of features, but also ensures that models can interpret and learn from the data effectively.

**Scaling** is applied to numerical features to align their value ranges, which is especially important for models sensitive to feature magnitude. The features `NearbyGeneralFacilities`, `NearbyPublicOffices`, `NearbyUniversities`, `BasementParkingSpaces`, `InApartmentFacilities`, and `SizeSqft` exhibit varying distributions and contain outliers, particularly the `SizeSqft` column. To address this, **Robust Scaler** is used, as it leverages the **median** and **interquartile range** to reduce the impact of extreme values. This makes it well-suited for **skewed data** and ensures fair representation across all scaled features.

**Encoding** transforms categorical variables into numerical format. **One-Hot Encoding** is used for **`HallwayType`** and **`SubwayStation`**, **nominal features** with few categories, to avoid introducing false ordinal relationships. This technique creates a binary column for each category, and one category is dropped (**k-1**) to prevent **multicollinearity**. **Ordinal Encoding** is applied to **`TimeToSubway`**, where the order of categories reflects increasing **proximity to a subway station**. A **custom mapping** ensures the model captures this meaningful order. These transformations are implemented using a **ColumnTransformer**, ensuring **consistency** and preserving important **structural relationships** in the dataset during preprocessing.


```python
# DATA TRANSFORMATION

# Features for scaling and encoding
scaling_features = ['NearbyGeneralFacilities', 'NearbyPublicOffices', 'NearbyUniversities', 'BasementParkingSpaces', 
                    'InApartmentFacilities', 'SizeSqft']
onehot_features = ['HallwayType', 'SubwayStation']
ordinal_features = ['TimeToSubway']

# Mapping for ordinal encoder 'TimeToSubway'
ordinal_map_timetosubway = ['no_subway_station_nearby', '15min~20min', '10min~15min', '5min~10min', '0~5min']

# Preprocessor for scaling and encoding
preprocessor = ColumnTransformer(
    transformers=[
        ('scaler', RobustScaler(), scaling_features),
        ('onehot', OneHotEncoder(drop='first'), onehot_features),
        ('ordinal', OrdinalEncoder(categories=ordinal_map_timetosubway), ordinal_features)
    ],
    remainder='passthrough'
)

preprocessor
```

#### **II.4.2 Feature Engineering**

**Feature Engineering**, as part of **data preprocessing**, aims to improve model performance by transforming, modifying, or organizing features in ways that make patterns in the data more accessible to machine learning algorithms. One common technique in feature engineering is **binning**, which groups continuous numerical variables into **discrete, ordered categories**. Binning simplifies complex data distributions, reduces the impact of noise and outliers, and can enhance the model’s ability to **identify patterns**, especially for algorithms that benefit from categorical or ordinal inputs.

In this context, `YearBuilt` and `SizeSqft` are binned into meaningful categories. The `YearBuilt` feature is segmented into four intervals: **Very Old**, **Old**, **Modern**, and **New**, based on key construction period trends, capturing the evolution of building standards over time. Similarly, `SizeSqft` is categorized into **Small**, **Medium**, **Large**, and **Very Large** based on property size ranges. These bins reflect practical groupings that help the model distinguish properties by age and size more effectively, while also reducing sensitivity to outliers and skewed distributions. 

Since both `YearBuilt` and `SizeSqft` have an inherent order after binning, **ordinal encoding** is applied to convert these categorical bins into numeric values that preserve their natural ranking, allowing algorithms to interpret and leverage the progression between categories. By converting continuous variables into interpretable ordinal categories, binning supports the model in learning more generalized and robust patterns, ultimately enhancing prediction accuracy and model interpretability.

```python
# Check the Descriptive Statistics of 'YearBuilt' and 'SizeSqft'
df[['YearBuilt', 'SizeSqft']].describe().T
```

```python
# BINNING 'YearBuilt' AND 'SizeSqft'

# a. Define bin edges and labels for 'YearBuilt'
year_bins = [1977, 1987, 1997, 2006, 2015]
year_labels = ['Very Old', 'Old', 'Modern', 'New']

# b. Define bin edges and labels for 'SizeSqft'
size_bins = [134, 685, 1236, 1787, 2337]
size_labels = ['Small', 'Medium', 'Large', 'Very Large']

# Create new binned columns
df['YearBuilt_Binned'] = pd.cut(df['YearBuilt'], bins=year_bins, labels=year_labels)
df['SizeSqft_Binned'] = pd.cut(df['SizeSqft'], bins=size_bins, labels=size_labels)

# Check the result
df[['YearBuilt', 'YearBuilt_Binned', 'SizeSqft', 'SizeSqft_Binned']].sample(10)
```

```python
# UPDATE PREPROCESSOR

# Define ordinal encoding categories for the binned features
ordinal_map_yearbuilt = ['Very Old', 'Old', 'Modern', 'New']
ordinal_map_sizesqft = ['Small', 'Medium', 'Large', 'Very Large']

# Update ordinal features
ordinal_features = ['TimeToSubway', 'YearBuilt_Binned', 'SizeSqft_Binned'] # add 'YearBuilt_Binned' and 'SizeSqft_Binned'

# Update the preprocessor
preprocessor = ColumnTransformer(
    transformers=[
        ('scaler', RobustScaler(), scaling_features),
        ('onehot', OneHotEncoder(drop='first'), onehot_features),
        # Update Ordinal encoder
        ('ordinal', OrdinalEncoder(categories=[ordinal_map_timetosubway, ordinal_map_yearbuilt, ordinal_map_sizesqft]),
         ordinal_features)
    ],
    remainder='passthrough'
)

preprocessor
```

### **III.1  Benchmark Model**

**Benchmarking** refers to the systematic comparison of multiple algorithms under **identical preprocessing and evaluation conditions** to establish a **baseline performance**. In the context of **Daegu apartment sale‐price prediction**, benchmarking quantifies each model’s **predictive ability** before any **hyperparameter tuning**, providing insight into **algorithmic strengths and weaknesses** and guiding the selection of **candidate models** for deeper optimization.

Ten **regression algorithms** were embedded in a **uniform end-to-end pipeline**, with consistent **data preprocessing**, including **scaling**, **encoding**, and **binning**, applied beforehand. This ensures that differences in predictive performance stem purely from **model characteristics**, not from variations in data preparation. Each model was **trained on the training set** and **evaluated on the unseen test set**, allowing for fair comparisons under identical test conditions.

The following **evaluation metrics** were computed to assess performance across multiple error dimensions:

- **Mean Absolute Error (MAE)** and **Mean Absolute Percentage Error (MAPE)** for interpretable average deviation in **absolute** and **relative** terms  
- **Mean Squared Error (MSE)** and **Root Mean Squared Error (RMSE)** to emphasize **larger residuals**  
- **Root Mean Squared Logarithmic Error (RMSLE)** and **Root Mean Squared Percentage Error (RMSPE)** capture sensitivity to **log-scale** and **percentage-based** errors  
- **R² (coefficient of determination)** evaluates how well the model **explains variance** in the target variable  

The selected models span from **linear regressors** (**Multiple Linear**, **Lasso**, **Ridge**) to **non-parametric learners** (**K-Nearest Neighbors**, **Decision Tree**), and up to **ensemble-based models** of increasing complexity: **Bagging**, **Random Forest**, and boosting methods such as **Gradient Boosting**, **XGBoost**, and **CatBoost**. #### **III.2.1 Hyperparameter Tuning**

**Hyperparameter tuning** is the process of optimizing the external configuration settings of a machine learning model, known as **hyperparameters**, which are not learned from the training data but instead control how the learning algorithm operates. Examples include **`alpha`** in regularized regression models, **`n_neighbors`** in KNN, or **`n_estimators`** and **`max_depth`** in ensemble models like Random Forest or Gradient Boosting.

Two widely used techniques for hyperparameter tuning are **grid search** and **randomized search**:

- **Grid search** exhaustively evaluates all combinations in a predefined parameter grid. It offers thoroughness but becomes **computationally expensive** and **time-consuming** as the number of parameters grows.

- **Randomized search** samples a fixed number of combinations from the parameter space. It enables **faster exploration** of a **broader value range** with reduced computation, though it does **not guarantee** the absolute best result.

In this project, **randomized search** is used due to its balance of **efficiency**, **coverage**, and **practical effectiveness**, particularly when only a few parameters have substantial impact on performance.

Moreover, to ensure robust evaluation during tuning, **cross-validation** is used. Instead of relying on a single train-test split, the data is divided into **5-folds**, and the model is trained and validated across multiple iterations. This process helps mitigate overfitting and ensures that the selected hyperparameters generalize well across different data subsets.

In practice, **hyperparameter tuning is essential** for most models, especially complex ones like **ensemble methods**, to reach peak performance. While linear regression has no tunable hyperparameters, models like **Lasso**, **Ridge**, **KNN**, **Decision Tree**, and **Ensemble algorithms** benefit significantly from careful tuning. Proper tuning not only improves accuracy but also contributes to **stability**, **efficiency**, and **interpretability** when deployed in real-world settings.

#### **III.2.2 Best Model Selection**

After completing **hyperparameter tuning** for ten regression models, the next step is to determine the **best-performing model**. This selection is critical, as models vary in how well they adapt to the data, even after tuning. A well-chosen model not only **fits the training data effectively** but also **generalizes well to unseen data** in real-world applications.

This project prioritizes two evaluation metrics: **Mean Absolute Error (MAE)** and **Mean Absolute Percentage Error (MAPE)**, due to their clarity, reliability, and practical significance. **MAE** offers a straightforward interpretation by quantifying the average prediction error in the same units as the target variable, making it especially relevant in domains like real estate where cost differences are critical. Unlike squared-error metrics such as RMSE, MAE does not disproportionately penalize large errors, which makes it **less sensitive to outliers** and more representative of typical performance.

**MAPE**, on the other hand, expresses errors as **percentages relative to actual values**, allowing for **scale-independent evaluation**. This makes it easier to understand how large the errors are in relation to the property values themselves, which is crucial in datasets with wide-ranging target values. It also facilitates **model comparison across different regions or markets** with different price levels, something RMSE and R² cannot directly support.

While metrics like **RMSE** emphasize large errors and **R²** explains variance without directly reflecting prediction accuracy, they are less interpretable for business stakeholders. In contrast, **MAE** and **MAPE** provide **actionable and intuitive feedback** that better aligns with real-world decision-making, budgeting, and risk assessment. For these reasons, they are more suitable for evaluating performance in this housing price prediction task.

#### **a. What is XGBoost?**

**XGBoost (Extreme Gradient Boosting)** is an **optimized and advanced implementation** of the **gradient boosting framework**, designed to deliver **high performance**, **scalability**, and **robust predictive accuracy**. While it shares the same foundational principle as traditional gradient boosting, which is building an ensemble of **weak learners** (typically decision trees) in a **sequential** manner to minimize residual errors, XGBoost introduces multiple enhancements that make it significantly more **efficient**, **powerful**, and **suitable for large-scale problems**.

Each new tree in XGBoost is trained to correct the **errors** of the current ensemble by minimizing a differentiable loss function, progressively improving the model’s predictions. However, unlike basic gradient boosting, XGBoost incorporates:

- **Regularization** (both **L1** and **L2**) to prevent overfitting by penalizing model complexity.
- **Advanced tree pruning** based on a **gamma** threshold, allowing the algorithm to eliminate branches that do not meaningfully reduce the loss.
- **Parallelized and distributed training**, making use of multi-core hardware and distributed systems to **dramatically speed up computation**.
- **Second-order optimization**, using both gradients and Hessians for more precise and stable updates.
- **Native support for missing values** through a sparsity-aware algorithm that learns optimal split directions for missing data.
- **Cache-aware and memory-efficient computation** to handle large datasets with better hardware utilization.

These innovations make XGBoost a **highly optimized, scalable, and production-ready** boosting algorithm, often the go-to choice for competitive modeling tasks and real-world deployment.


#### **b. How XGBoost Works?**

XGBoost builds its predictive power through an iterative process rooted in the **gradient boosting** framework, but enhances it with sophisticated techniques for **efficiency**, **regularization**, and **scalability**. 

Here's a breakdown of its internal workings:

1. **Model Initialization**

    XGBoost begins with a simple initial prediction for all training samples, typically the **mean** of the target variable in regression tasks. This serves as the baseline upon which subsequent models iteratively improve.

2. **Boosting through Iterative Tree Building**

    The core of XGBoost lies in **building decision trees sequentially**, where each new tree is trained to capture the **residual errors** (or pseudo-residuals) made by the ensemble up to that point. These residuals represent the **gradient** (i.e., first-order derivative) of the loss function with respect to the model's current predictions.

    For each boosting round:

    - A new decision tree is trained to predict these gradients using only a **subset of samples (`subsample`) and features (`colsample_bytree`)**. This introduces randomness, which improves generalization and reduces overfitting.
      
    - The tree’s output is **scaled by a learning rate (`learning_rate`)**, a technique known as **shrinkage**, to control how much each tree influences the final prediction. This slows down learning in favor of greater stability.

    - Once added to the model, the ensemble predictions are updated, and the process repeats with new gradients calculated based on the updated predictions.

    - Tree **pruning** is applied during construction using the **`gamma`** parameter. If a branch doesn’t yield a significant reduction in the loss function, it is pruned early. This results in **compact and efficient trees** that focus only on meaningful splits.

3. **Regularization for Generalization**

    - Unlike traditional gradient boosting, XGBoost explicitly includes **regularization terms** in its objective function:

      - **L1 regularization** encourages sparsity in leaf weights (some outputs may be zero).
      - **L2 regularization** penalizes large leaf weights to reduce variance and prevent overfitting.

    - Other important regularization-related controls include:

      - **`min_child_weight`**, which ensures that nodes are only split if they contain a sufficient number of training instances (or their total weight). This prevents the model from creating overly specific rules that only apply to a small subset of the data.

    - These mechanisms allow XGBoost to maintain **high accuracy** while **resisting overfitting**, even on complex or noisy datasets.

4. **Computational Efficiency and Parallelism**

    XGBoost is designed for **high-performance training**:

    - It uses a **column-block data structure** that optimizes memory access and enables **feature-wise parallelism**. Different features can be processed in parallel during split finding.

    - It supports **histogram-based algorithms** for faster and more scalable split finding, especially effective for large datasets.

    - **Parallel and distributed training** make it suitable for multi-core CPUs and even distributed computing environments, accelerating learning on big data.

    - It also handles **sparse inputs and missing values** efficiently by learning optimal default directions for missing data during tree construction.

It does not capture bias or absolute error magnitude, so it should be complemented with error-based metrics.          |

#### **III.3.1 Residual Analysis**

**Residual analysis** is a fundamental diagnostic technique in regression modeling that involves examining the differences between observed and predicted values, known as **residuals**. A residual is calculated as the **actual value minus the predicted value** for each observation. This analysis is essential for evaluating how well the model captures the underlying data structure and whether its predictive behavior is consistent across the entire range of target values.

Although tree-based models like **XGBoost** do not rely on traditional linear regression assumptions, such as **linearity**, **normality**, or **homoscedasticity**, residual analysis remains highly relevant. It can uncover critical issues such as **non-linearity**, **heteroscedasticity**, or the presence of **systematic prediction errors** in specific value ranges. These insights are especially useful for identifying weaknesses in the model, guiding further **feature engineering**, or flagging areas where model performance may require improvement.

In this project, residual analysis is conducted using three key visual diagnostics. A **distribution plot** examines the overall shape and symmetry of residuals, helping to detect **skewness** or **outliers**. A **Q-Q plot** visually assesses how closely the residuals follow a **normal distribution**, which, while not required by XGBoost, can still indicate whether errors are evenly spread. Finally, a **residuals vs. predicted values plot** checks for visible patterns or **unequal variance** across predictions, signs of potential model misspecification. Together, these tools provide a deeper understanding of model behavior and help validate its **reliability** in practical applications.


```python
# RESIDUAL VISUALIZATION

# Calculate residuals
residuals = y_test - y_pred_final

# 1. Plot the distribution of residuals
sns.displot(residuals, kde=True, color='darkcyan', height=5, aspect=6/4)
plt.title('Distribution of Residuals')
plt.xlabel('Residual Value')
plt.ylabel('Frequency')
plt.show()

# 2. Q-Q Plot of Residuals
plt.figure(figsize=(8, 5))
stats.probplot(residuals, dist="norm", plot=plt)
plt.title('Q-Q Plot of Residuals')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')
plt.grid(True)
plt.gca().get_lines()[0].set_color('darkcyan')
plt.gca().get_lines()[1].set_color('chocolate')
plt.show()

# 3. Residuals vs. Predicted Values
plt.figure(figsize=(8, 5))
sns.scatterplot(x=y_pred_final, y=residuals, color='darkcyan')
plt.axhline(0, color='chocolate', linestyle='--')
plt.xlabel('Predicted Sale Price')
plt.ylabel('Residual (Actual - Predicted)')
plt.title('Residuals vs. Predicted Sale Price')
plt.grid(True)
plt.show()
```

####    <u>**Summary of Residual Analysis**</u>

**Residual analysis**, even for tree-based models such as XGBoost, remains a critical step for **uncovering hidden patterns in prediction errors**. The following provides a **summary** of the three diagnostic plots.

#### **a. Distribution of Residuals**
1. **Balanced Core**

3. **Impact of Skew and Tails**  
  This asymmetry and heavy right tail inflate overall error metrics and signal that certain segments (likely high-value or unusual apartments) are not captured as reliably.

#### **b. Q–Q Plot of Residuals**
1. **Central Alignment**  
  Most points fall on the 45° line in the middle quantiles, confirming that the core residuals approximate a normal distribution and support valid inference for typical cases.
2. **Heavy Tails**  
  Systematic upward deviations at both extremes, especially the upper tail, indicate more extreme under-predictions than a Gaussian model would allow, echoing the right-skew in the histogram.

  These heavier-than-normal tails mean that prediction intervals based on a pure normal assumption will be too narrow for the most expensive apartments, underestimating uncertainty where it matters most.

#### **c. Residuals vs. Predicted Values**

  Residuals remain tightly clustered around zero across low and mid-range predictions, showing no systematic bias at those levels.
2. **Fan-Shaped Heteroscedasticity**

  A handful of extreme residuals at the top end further highlight cases where the model struggles, potentially pulling global metrics without shifting the average error.

**Overall**, the **XGBoost model captures the central trend very well**, but **exhibits mild right-skew** and **increasing variance at higher prices**, which are key signs that prediction uncertainty is under-estimated for the most expensive apartments.

####    <u>**Summary of Feature Importance Analysis**</u>

Understanding the distinction between **Gain-based Feature Importance** and **SHAP values** is critical for developing both **accurate models** and **interpretable insights**. Though both quantify feature contributions, they do so from fundamentally different perspectives, offering complementary insights.

1. **Global Aggregation vs. Local Attribution**  
   - **Gain** measures average accuracy improvement when a feature is used in splits, providing a **global ranking** of feature relevance.  
   - **SHAP** quantifies the precise contribution of each feature to **individual predictions**, capturing variability across samples.  
   - **Key distinction:** Gain reflects the model’s structural reliance, SHAP reflects output behavior.

2. **Feature Interactions and Directionality**  
   - Gain ignores the direction of influence and interactions between features.  
   - SHAP captures both **magnitude and direction**, revealing how features increase or decrease predictions and uncovering interaction effects.

3. **Interpretability Scope and Depth**  
   - Gain offers a high-level overview, ideal for identifying key predictors but lacks instance-specific context.  
   - SHAP provides detailed, instance-level explanations, essential for personalized insights and decision-making.

4. **Sensitivity to Feature Distribution**  
   - Gain can be biased toward features with many split points or early splits.  
   - SHAP fairly distributes importance among correlated and interacting features, enhancing robustness.

5. **Complementary Interpretive Strengths**  
   - Gain answers: *“Which features drive overall model performance?”*  
   - SHAP answers: *“How does each feature affect specific predictions?”*  
   - Combined, they deliver a full picture of feature influence and behavior.

<br>

#### **Reasons for Differences in Gain vs. SHAP Feature Importance Rankings**

| **Cause of Ranking Differences**         | **Explanation**                                                                                                                                                      | **Application (Daegu Apartment Price Model)**                                                                                   |
|-----------------------------------------|------------------------------------------------------------------------------------------------------------------------------------------------------------------|---------------------------------------------------------------------------------------------------------------------------------|
| **Global vs. Local Focus**                | Gain aggregates importance from splits across the whole model, while SHAP captures feature impact on individual predictions, including consistent linear effects. | `SizeSqft` ranks higher in SHAP due to its steady linear influence, despite fewer high-gain splits in the model structure.       |
| **Bias Toward Feature Split Frequency**  | Gain favors features with many split opportunities (e.g., continuous or binary features). SHAP allocates importance based on actual contribution.                | `HallwayType_terraced` shows inflated Gain importance due to split bias, but SHAP balances its influence relative to others.    |
| **Handling Correlations and Interactions** | SHAP distributes importance among correlated features by accounting for interactions; Gain may attribute importance predominantly to one representative feature.  | SHAP splits credit between `SizeSqft` and its binned version, whereas Gain may emphasize only one.                              |
| **Directionality and Effect Sign**        | Gain measures only error reduction magnitude, ignoring whether features increase or decrease predictions. SHAP reflects both direction and magnitude.            | `YearBuilt` has high Gain but SHAP reveals its negative effect on price for older buildings, enhancing interpretability.         |
| **Consistent vs. Sporadic Effects**       | SHAP captures subtle, consistent contributions across many samples; Gain highlights features causing fewer, larger error reductions.                            | `NearbyGeneralFacilities` ranks higher in SHAP due to steady influence, despite limited high-gain splits in the model.          |
